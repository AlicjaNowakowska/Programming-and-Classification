### zad 13

```python
#Construct a function that returns a string of maximal length from a set of strings.
#Note the the maximal do not have to be unique.
def max_string(lista):
    maximum=len(max(['safdg','a','bbbbb','sdf'], key=len))
    return [i for i in lista if len(i)==maximum]
```

```python
max_string(['safdg','a','bbbbb','sdf'])
```

### zad 14

```python
"""Construct a function that construct a list that keeps alternately consecutive elements from two argument lists of equal length. For example for a=[a1,a2,a3] and
a=[b1,b2,b3] a correct output is [a1,b1,a2,b2,a3,b3] .
"""
def zad14(a,b):
    new_list=[]
    for i in range(len(a)):
        new_list.append(a[i])
        new_list.append(b[i])
    return new_list
```

```python
zad14([1,2,3,4],[5,6,7,8])
```

### zad 15

```python
"""A list contains both integers and strings. Construct a function that sorts a given list
in such a way that at the beginning of the resulting list one gets string in alphabetical
order followed by growing integers."""
def sort(lista):
    ints=[i for i in lista if type(i)==int]
    strings=[i for i in lista if type(i)==str]
    ints.sort()
    strings.sort()
    return strings+ints
```

```python
sort([-1,2,'nana','a',-5,8,'c'])
```

### zad 16

```python
"""Construct a dictionary that assigns each user (string) his address (string). Sort
and list this dictionary by the addresses. Use lambda function."""
def zad16(users,adresses):
    slownik={}
    for i in range(0,len(users)):
        slownik[users[i]]=adresses[i]
    return sorted(slownik.items(), key = lambda kv:(kv[1], kv[0]))
```

```python
zad16(['a','b','k','e'],['5','3','2','8'])
```
